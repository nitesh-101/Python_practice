```python
with open('data.txt', ) as fh:
    words=fh.read().split() 
    unique_words = set(words)
    count = len(unique_words)
    print("Unique words:", unique_words)
    print("Total unique words:", count)
```

```python
with open('data.txt') as fh:
    words=fh.read().split()
    word_count={}  
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
print(word_count)
```

```python
# JSON: 
import json
data = {
    "name": "Nitesh",
    "age": 24,
    "course": "MCA"
}
with open('json1.json', 'w') as fh:
    json.dump(data, fh)
print(data)
```

```python
with open('json1.json', 'r') as fh:
    # json.dump(data, fh)
    print(json.load(fh))
```

```python
# CSV File
import csv
with open('data.csv', 'w') as file:
    writer=csv.writer(file)
    writer.writerow(['Name', 'Age', 'Course'])
    writer.writerow(['Nitesh','24','Mca'])
```

```python
# To remove new line
with open('data.csv', 'w', newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Name', 'Age', 'Course'])
    writer.writerow(['Nitesh','24','Mca'])
```

```python
# Add multiple rows
import csv
rows=[
    ['Name', 'Age'],
    ['Aman', 21],
    ['Ravi', 23]
]
with open('data.csv', 'w', newline='') as file:
    writer=csv.writer(file)
    writer.writerows(rows)
```

```python
# Writes data using dictionary
with open('data.csv', 'w', newline='') as file:
    fieldnames=["Name","Age"]
    writer=csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"Name":"Nitesh", "Age":23})
    writer.writerow({"Name":"Avishek", "Age":23})
```

```python
with open("data.csv", "r") as file:
    reader=csv.DictReader(file)
    print(reader) # -> Object address
    for row in reader:
        print(row)
```
