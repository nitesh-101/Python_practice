### Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,50,10)
y=np.random.randint(1,50,10)
print(x, y, sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color='red')
```

```python
plt.plot(x,y, color='red', linestyle='dotted')
plt.show()  # Remove the object address
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.plot(x,y, color='red', linestyle='dotted')
plt.show() 
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  # To dsiplay datapoints we use xticks and yticks
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='dotted')
plt.show() 
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-')
plt.show() 
```

```python
# - -> solid(default)
# -- -> dashed
# -. -> dash-DisplayObject
# : -> dotted
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-.')
plt.show() 
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9) #alpha -> used to increase or decrease intensity
plt.show() 
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9, marker='*')   #marker used to mark the datapoints 
plt.show() 
```

```python
# Change internal color of marker
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9, marker='*', markerfacecolor='y')    
plt.show() 
```

```python
# Change internal color of marker
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9, marker='*', markerfacecolor='y', markersize=10)    
plt.show() 
```

```python
# Edge Color
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9, marker='*', markerfacecolor='y', markersize=10, markeredgecolor='blue')    
plt.show() 
```

```python
# Width
plt.title('Random Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.xticks(x)  
plt.yticks(y)
plt.plot(x,y, color='red', linestyle='-', alpha=0.9, marker='*', markerfacecolor='y', markersize=10, markeredgecolor='blue',
         linewidth=3)    
plt.show() 
```

```python
x=np.arange(1,11)
y=np.sin(x)
z=np.cos(x)
print(x,y,z, sep='\n\n')
```

```python
plt.plot(x,y, label='SINQ', color='red')
plt.plot(x,z, label='COSQ')
plt.title('SINQ & COSQ', color='green')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.xticks(y)  
plt.yticks(z)
plt.legend()  # To view the label
plt.show()
```

```python
# Height and Width
plt.figure(figsize=(8,4))  # First Width then Height (8,4)
plt.plot(x,y, label='SINQ', color='red')
plt.plot(x,z, label='COSQ')
plt.title('SINQ & COSQ', color='green')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.xticks()  
plt.yticks()
plt.legend()  # To view the label
plt.show()
```

```python
# Height and Width
plt.figure(figsize=(8,4))  # First Width then Height (8,4)
plt.subplot(1,2,1)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='red')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.legend()

plt.subplot(1,2,2)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='green')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points2')
plt.ylabel('Associate Values2')

plt.legend()  # To view the label
plt.show()
```

```python
plt.figure(figsize=(8,8))  # First Width then Height (8,4)
plt.subplot(3,3,1)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='red')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.legend(loc="upper left")

plt.subplot(3,3,5)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='green')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points2')
plt.ylabel('Associate Values2')
plt.legend(loc='upper right')

plt.subplot(3,3,9)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='green')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points2')
plt.ylabel('Associate Values2')
plt.legend(loc='upper right')  # To view the label

plt.show()
```

```python
plt.figure(figsize=(8,4))  # First Width then Height (8,4)
plt.subplot(1,2,1)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='red')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points')
plt.ylabel('Associate Values')
plt.grid()
plt.legend(loc='upper right')

plt.subplot(1,2,2)
plt.title('SINQ & COSQ', color='green')
plt.plot(x,y, label='SINQ', color='green')
plt.plot(x,z, label='COSQ')
plt.xlabel('Data Points2')
plt.ylabel('Associate Values2')
plt.subplots_adjust(wspace=0.8)  # To overcome overlap issues
plt.grid()
plt.legend(loc='upper right')  # To view the label

plt.show()
```

### Scatter Plot

```python
x=np.arange(1, 11)
y=np.random.randint(1,100,10)
print(x,y, sep='\n')
```

```python
plt.scatter(x,y)
plt.show()
```

```python
company=['a','b','c','d']
revenue=[12,34,45,54]
plt.scatter(company,revenue)
```

```python
plt.bar(company,revenue, color=['blue', 'orange', 'green', 'red'])
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.scatter(
    x=x, y=y, marker='v',
    edgecolor='y', alpha=0.9,   
)
```

```python
#Barh plot used for long labels or horizontal plots
company=['aaaaaavbbgfh','b','c','d']
revenue=[12,34,45,54]
plt.barh(company,revenue, color=['blue', 'orange', 'green', 'red'])
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
#Histogram plot  used for continuous data
plt.hist(revenue)
plt.title("Histogram Plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
#Pie plot used to get the % of distribution
plt.pie(revenue, labels=company, autopct="%.3f")
plt.title("Pie Plot")
plt.show()
```

```python
# Line plot
company=['A','B','C','D']
revenue=[120,34,35,54]
plt.plot(revenue, company)
plt.title("Line Plot")
plt.show()
```

```python
# Box Plot used for outliers detection
plt.boxplot(revenue)
plt.title("Box plot")
plt.show()
```
