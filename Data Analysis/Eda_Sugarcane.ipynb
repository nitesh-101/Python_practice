```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("List of Countries by Sugarcane Production.csv")
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Production (Tons)']=df['Production (Tons)'].str.replace('.','')
df.head()
```

```python
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace('.','')
df.head()
```

```python
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace(',','')
df.head()
```

```python
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].str.replace('.','').str.replace(',','')
df.head()
```

```python
df['Acreage (Hectare)']=df['Acreage (Hectare)'].str.replace('.','')
df.head()
```

```python
#Rename
df.rename(
    columns={
        'Production (Tons)': 'Production_Tons',
        'Production per Person (Kg)': 'Production_per_Person_Kg',
        'Acreage (Hectare)': 'Acreage_Hectare',
        'Yield (Kg / Hectare)': 'Yield_Kg_Hectare'
        },
        inplace=True
)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.Acreage_Hectare.isnull()]
```

```python
# Delete record 99
df.dropna()
```

```python
df.dropna(inplace=True)
df.head(100)
```

```python
df.reset_index(drop=True, inplace=True)
df.head(100)
```

```python
# How many countries produce sugarcane from each continent?
df.Continent.value_counts()
```

```python
cont = df.Continent.value_counts()
cont
plt.figure(figsize=(10,6))
plt.title('Top Sugarcane Production Countries')
plt.bar(cont.keys(), cont)
plt.show
```

```python
# Another way
df.Continent.value_counts().plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.title('Continent-wise Count')
plt.show()
```

```python
# Which Country produces Maximum Sugarcane?
df.loc[df['Production_Tons'].idxmax(), 'Country']
```

```python
new_df = df.loc[:, ["Country", "Production_Tons"]].sort_values(by="Production_Tons", ascending=False)
new_df.head()
```

```python
# Top 5 Maximum Sugar Production Countries %
new_df["Production(Tons)_Percentage"] = (new_df["Production_Tons"]*100)/sum(new_df["Production_Tons"])
new_df.head()
```

```python
import matplotlib.pyplot as plt

top5 = new_df.head(5)

plt.figure(figsize=(8, 5))
plt.pie(
    top5['Production(Tons)_Percentage'],
    labels=top5['Country'],
    autopct='%0.1f%%'
)
plt.title('Top 5 Sugar Production Countries (%)')
plt.show()
```
