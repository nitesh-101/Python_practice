```python
# Merge and Concat
# Concatination
data1={
    'Name':['A', 'B', 'C'],
    'age':[23,45,67],
    'salary':[1000,2000,3000]
}
data2={
    'Name':['A', 'BB', 'C'],
    'age':[23,54,67],
    'salary':[1000,2500,3000]
}
```

```python
new_df1=pd.DataFrame(data1)
new_df2=pd.DataFrame(data2)
```

```python
new_df1.head()
```

```python
new_df2.head()
```

```python
pd.concat([new_df1, new_df2])
```

```python
pd.concat([new_df1, new_df2], ignore_index=True)
```

```python
# Column wise adding data so use axis=1 and if by row then axis=0 and by default it is zer0
pd.concat([new_df1, new_df2], ignore_index=True, axis=1)
```

```python
new_df=pd.concat([new_df1, new_df2], ignore_index=True, axis=0)
new_df.head(6)
```

```python
# To delete (row by row data is deleted)
new_df.drop_duplicates(inplace=True)
new_df.head()
```

```python
# Merge
pd.merge(new_df1, new_df2)
new_df2.head()
```

```python
# Use sql joins
pd.merge(new_df1, new_df2, on='age', how='outer')
```
