```python
no1=df['total_bill'][:31]
no2=df['tip'][:31]

plt.figure(figsize=(8,8))
plt.subplot(3,3,1)
plt.title('TIPS', color='green')
plt.bar(no1, no2, label='Tip vs Total Bill', color='red')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend(loc="upper left")

plt.subplot(3,3,3)
plt.title('TIPS', color='green')
plt.scatter(no1, no2, label='Tip vs Total Bill', color='red')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend(loc="upper left")

plt.subplot(3,3,5)
plt.title('TIPS', color='green')
plt.plot(no1, no2, label='Tip vs Total_bill', color='red')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend(loc="upper left")

plt.subplot(3,3,7)
plt.title('TIPS', color='green')
plt.hist(no2)
plt.xlabel('total_bill')
plt.ylabel('tip')

plt.subplot(3,3,9)
plt.title('total_bill', color='green')
plt.pie(no1, autopct="%.1f%%")

plt.show()
```
